```python
import plotly.graph_objects as go
import pandas as pd
import plotly as py
from chart_studio.plotly import plot, iplot
import numpy as np
```

```python
aedes=pd.read_csv("aegypti_albopictus.csv")
```

```python
aedes.head()
```

```python
aedes.shape
```

```python
aedes.Y.value_counts()
```

```python
 aedes.isnull().any()
```

```python

aedes = aedes[pd.notnull(aedes["YEAR"])]
```

```python
 aedes.isnull().any()
```

```python
aedes.head()
```

```python
import numpy as np
```

```python
fig = go.Figure(data=go.Scattergeo(
        lon = aedes['X'],
        lat = aedes['Y'],
        mode = 'markers',
        marker=dict(color=np.where(aedes["VECTOR"]=="Aedes aegypti", 'red', 'green')),
        text= aedes['VECTOR']))

fig.update_layout(
        title = 'Global distribution of Aedes vector species',
        geo = dict(
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    )
fig.show()




```
